# ANALYSIS

GROUPING BY BORO

```python
# BROOKLYN
# Step 1: Filter the dataset to only Brooklyn, then count offenses by race
brooklyn_df =df[df["BORO"] == "BROOKLYN"]  # select only the rows where the borough is Brooklyn
race_counts_brooklyn = brooklyn_df.groupby("RACE").size()  # group the Brooklyn rows by race and count how many summonses fall in each race group
race_counts_brooklyn = race_counts_brooklyn.reset_index()  # convert the grouped Series back into a regular DataFrame
race_counts_brooklyn.columns = ["RACE", "COUNT"]  # rename the columns for clarity
race_counts_brooklyn = race_counts_brooklyn.sort_values(by="COUNT", ascending=False)  # sort the races from most offenses to fewest
print(race_counts_brooklyn)  # display the resulting table showing race with the most offenses in Brooklyn
```

```python
# Step 1: Filter Brooklyn data to only Black, Black Hispanic, and White Hispanic, then pivot offense vs race
selected_races = ["BLACK", "BLACK HISPANIC", "WHITE HISPANIC"]  # define the list of race categories to keep
brooklyn_filtered = brooklyn_df[brooklyn_df["RACE"].isin(selected_races)]  # keep only rows where race matches one of the selected categories
brooklyn_offense_by_race_filtered = brooklyn_filtered.pivot_table(index="OFFENSE_DESCRIPTION", columns="RACE", values="SUMMONS_KEY", aggfunc="count", fill_value=0)  # reshape data so each selected race is a column, each offense is a row, and cells show the count of summonses
print(brooklyn_offense_by_race_filtered.head(20))  # display the first 20 rows of the filtered side-by-side comparison table
```
