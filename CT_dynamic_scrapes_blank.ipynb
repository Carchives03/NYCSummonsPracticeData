Get all the boroughs

```python
# list of boros
boros = ["Manhattan", "Brooklyn", "Queens", "Bronx", "Staten%20Island"]
```

```python
## slice boro
boros[2]
```

```python
## for loop thru boros
for boro in boros:
  print(f"This is {boro.upper()}")
```

```python
url_placeholder = "https://a816-health.nyc.gov/ABCEatsRestaurants/App/GetEntitiesByBoro/"
```

```python
# iteragte to scrape all boros
df_list = [] # empty list to append dfs

for boro in boros: #iterate through boros
  url = f"{url_placeholder}{boro}" # create url
  print(url) # flag where i am
  response = requests.get(url) # grab url
  df_list.append(pd.DataFrame(response.json())) # append to empty list

```

```python
len(df_list)
```

```python
df_list[3]
```

```python
df = pd.concat(df_list, ignore_index = True )
```
